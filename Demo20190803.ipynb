## Activation Function

### Threshold Function

```python
x = np.array([-1,1,2])
x > 0 
```

```python
(x > 0 ).astype(int)
```

```python
def threshold_function(x):
    y = x > 0
    return y.astype(int)

x = np.array([-1,1,2])
threshold_function(x)
```

### Sigmoid Function

```python
def sigmoid_function(x):
    return 1/ (1 + np.exp(-x))

x = np.array([-1,1,2])
sigmoid_function(x)
```

### Tangent Function

```python
def tangent_function(x):
    return (1 - np.exp(-2*x)) / (1 + np.exp(-2*x))

x = np.array([-1,1,2])
tangent_function(x)
```

### ReLU Function

```python
def relu_function(x):
    return np.maximum(0,x)

x = np.array([-1,1,2])
relu_function(x)
```

```python
def cross_entropy_err(y_hat, y):
    delta = 1e-8
    return -np.sum(y*np.log(y_hat + delta))
```

```python
def dfunc(f, x):
    h = 1e-4
    return (f(x+h) - f(x)) / (h)
```

```python
def tfunc(f,x,t):
    d= dfunc(f,x)
    y=f(x)-d*x
    return d*t+y
```

### Gradient Descent

```python
def gradient_descent(func, init_x, lr = 0.3, epochs = 100):
    x = init_x
    res = [x]
    for i in range(epochs):
        grad = dfunc(func,x)
        x = x - grad * lr
        res.append(x)
    return np.array(res)
```
